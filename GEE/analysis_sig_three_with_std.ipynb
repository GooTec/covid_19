```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

def standardize(col_dat):
#     col_dat = col_dat.dropna()
    std_col = ( col_dat - np.mean(col_dat) )/ np.std(col_dat)
    return std_col
```

```python
data = pd.read_csv('/data/member/twgoo/COVID-19/trajectory_analysis/oxford_globa.csv', index_col=0)
pre_country_code = set(data['CountryCode'])
print("Total number of countries :",len(pre_country_code))
```

```python
std_data = data.iloc[:,2:-2].apply(lambda x : standardize(x), axis=0)
data.iloc[:,2:-2] = std_data
```

```python
data = data.fillna(0)
```

```python
# fig = plt.figure(figsize=(20,20))
nrow = 10 
ncol = 4
fig, axes = plt.subplots(nrows=nrow, ncols=ncol, figsize=(20,20))
fig.subplots_adjust(hspace=0.15, wspace=0.1)
col_name = 'C8_International travel controls'
for i, country_code in enumerate(pre_country_code) : 
    policy_df = data[data['CountryCode'] == country_code]
    row,col = [i//4, i%4]
#     print(row,col)
    x = policy_df.iloc[:,1]
    y = policy_df[col_name].fillna(0)
    axes[row, col].set(xlabel='days from 2020-01-01', ylabel='col_value / max(col_value) * 100')
    axes[row, col].set_title(country_code, size=10)
    axes[row, col].plot(x, y/max(y)*100, '-', alpha = 0.8) 
    y = policy_df.iloc[:,-1].fillna(0)
    axes[row, col].plot(x, y/max(y)*100, '-', alpha = 0.8) 
#     print(row,col)
    count = 30  
    if col != 0 :
        axes[row, col].yaxis.set_visible(False)
    if col == 3 :
        ax2 = axes[row, col].twinx()
        ax2.set_ylabel("Cumulative confirmer")
        ax2.set_ylim(min(y), max(y)) 
    
    if row != 9 :
        axes[row,col].xaxis.set_visible(False) 
     
    
    while count < max(x) :  
        axes[row, col].axvline(count , linestyle='--', color='red', alpha=0.3)
        count += 30
plt.show()
```
